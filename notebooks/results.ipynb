```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_hrf = pd.read_csv('/home/eric/src/bptt-julia/Results/default/test/192/logs.csv', sep=',')

df_lin = pd.read_csv('/home/eric/src/bptt-julia/Results/default/test/194/logs.csv', sep=',')
```

```python
df_hrf.D_stsp.min()
```

```python
fig, ax = plt.subplots(2,2,figsize=(15,9))



ax[0,0].plot(df_hrf.index, df_hrf.Loss, label="Loss Hrf")
ax[0,0].plot(df_lin.index, df_lin.Loss, label="Loss Lin")
ax[0,0].set_title("Loss", fontsize=15)
ax[0,0].legend()

ax[0,1].plot(df_hrf.index, df_hrf.D_stsp, label="D_stsp Hrf")
ax[0,1].plot(df_lin.index, df_lin.D_stsp, label="D_stsp Lin")
ax[0,1].set_title("D_stsp", fontsize=15)
ax[0,1].legend()

ax[1,0].plot(df_hrf.index, df_hrf.D_H, label="D_H Hrf")
ax[1,0].plot(df_lin.index, df_lin.D_H, label="D_H Lin")
ax[1,0].set_title("D Hellinger", fontsize=15)
ax[1,0].legend()

ax[1,1].plot(df_hrf.index, df_hrf['20-step PE'], label="20-step PE Hrf")
ax[1,1].plot(df_lin.index, df_lin['20-step PE'], label="20-step PE Lin")
ax[1,1].set_title("20 step Predition Error", fontsize=15)
ax[1,0].legend()
```
